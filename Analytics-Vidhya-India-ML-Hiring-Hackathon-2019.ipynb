# Feature Engineering

```python
df['No of deliquent months max'] = df[['m1','m2','m3','m4','m5','m6','m7','m8','m9','m10','m11','m12']].max(axis=1)
```

```python
df['Credit Score-Max'] = df[['borrower_credit_score','co-borrower_credit_score']].max(axis=1)
```

```python
df['unpaid_principal_bal.BINS']=pd.cut(df['unpaid_principal_bal'],4,labels=['A','B','C','D']).astype(np.object)
```

```python
df['Debt to income.BINS']=pd.cut(df['debt_to_income_ratio'],3,labels=['Low','Medium','High']).astype(np.object)
```

```python
df['Value of Loan']=df['unpaid_principal_bal']/df['loan_to_value']
```

```python
df['loan_term.BINS']=pd.cut(df['loan_term'],3,labels=['A','B','C']).astype(np.object)
```

# Model building

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
#Uploading Test File
test = pd.read_csv("test.csv")
#Dropping UnRequired Attributes from test data:-
test= test.drop(['loan_id' ,'origination_date','first_payment_date','insurance_percent','insurance_type'],axis=1)
new = pd.read_csv("test.csv")
```

```python
from sklearn.preprocessing import LabelEncoder
lm = LabelEncoder()
a  = ['financial_institution']
for i in np.arange(len(a)):
    X1[a[i]] = lm.fit_transform(X1[a[i]])
    
for i in np.arange(len(a)):
    test[a[i]] = lm.fit_transform(test[a[i]])
```

```python
source = pd.get_dummies(X1['source'])
LoanPurpose = pd.get_dummies(X1['loan_purpose'])
```

```python
X1 = pd.concat([X1,LoanPurpose],axis = 1)
X1 = X1.drop(['source','loan_purpose'],axis=1)
```

```python
source1 = pd.get_dummies(test['source'])
LoanPu = pd.get_dummies(test['loan_purpose'])
```

```python
Test = pd.concat([test,LoanPu],axis = 1)
Test = Test.drop(['source','loan_purpose'],axis=1)
```

```python
from imblearn.over_sampling import ADASYN 
sm = ADASYN()
X, y = sm.fit_sample(X1, y)
X = pd.DataFrame(X, columns = X1.columns)
```

```python
model = RandomForestClassifier(random_state=42,max_depth= 10 ,min_samples_split= 5 ,n_estimators= 500,n_jobs=-1, min_samples_leaf=2,  criterion='entropy',class_weight= {0: 0.8, 1: 1.1})
model.fit(X,y)
y_pred = model.predict(Test)
```

```python
pd.DataFrame({'loan_id':new.loan_id,'m13':y_pred}).set_index('loan_id').to_csv('submit.csv')
pd.Series(model.feature_importances_,index=X1.columns).sort_values(ascending=False)
```
